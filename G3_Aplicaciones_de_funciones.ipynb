# Problema 1

Sean las funciones:

\begin{split}
r(x) & = -1,5x^{2} + 11,5x - 15\\
g(x) & = 1,58x^{4} - 19,17x^{3} + 80,92x^{2} - 139,33x + 85
\end{split}

<br>

Considerando que $\hspace{0.2cm}$ $0 \leq x \leq 6$, $\hspace{0.3cm}$  $-10 \leq y \leq 10$ $\hspace{0.2cm}$ realice lo siguiente:

1. Grafique ambas funciones considerando el dominio indicado. Sugerencia: utilice *plt.xlim()* y *plt.ylim()* para definir los límites de los ejes en el gráfico.
2. ¿Cuáles son los puntos de intersección entre las funciones? Utilice *fsolve* para encontrar las soluciones.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

def r(x):
    return -1.5 *x**2+11.5*x-15
def g(x):
    return 1.58*x**4-19.17*x**3+80.92*x**2-139.33*x+85

x = np.arange(0,10, 0.1)
plt.plot(x, r(x), label = 'r(x)')
plt.plot(x, g(x), label = 'g(x)')
plt.title('Nombre del gráfico')
plt.ylabel('Nombre eje y')
plt.xlabel('Nombre eje x')
plt.xlim(0,6)
plt.ylim(-10,10)
plt.legend()
plt.grid()


#opcional:
ax=plt.gca()
ax.spines[["left","bottom"]].set_position(("data",0))
ax.spines[["right","top"]].set_visible(False)


plt.show()
```

```python
import sympy as sp 
def r(x):
    return -1.5 *x**2+11.5*x-15
def g(x):
    return 1.58*x**4-19.17*x**3+80.92*x**2-139.33*x+85
#para encontrar puntos de interseccion, hay que resolver r(x) = g(x)
# por tanto en pyhton seria r(x) - g(x)

x=sp.Symbol("x")
ecuacion = r(x)-g(x) #la funcion es de grado 4 que tiene 4 soluciones
                      
solucion=sp.solve(ecuacion,x)
#print(solucion)
x1 = round(solucion[0],5)
x2 = round(solucion[1],5)
x3 = round(solucion[2],5)
x4 = round(solucion[3],5)

print(x1,x2,x3,x4)


import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,10, 0.1)
plt.plot(x, r(x), label = 'r(x)', color = 'blue')
plt.plot(x, g(x), label = 'g(x)', color = 'purple')
plt.plot(x1,r(x1), marker = '.', color = 'black')
plt.plot(x2,r(x2), marker = '.', color = 'black')
plt.plot(x3,r(x3), marker = '.', color = 'black')
plt.plot(x4,r(x4), marker = '.', color = 'black')
plt.title('Nombre del gráfico')
plt.ylabel('Nombre eje y')
plt.xlabel('Nombre eje x')
plt.xlim(0,6)
plt.ylim(-10,10)
plt.legend()
plt.grid()


#opcional:
ax=plt.gca()
ax.spines[["left","bottom"]].set_position(("data",0))
ax.spines[["right","top"]].set_visible(False)


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def w(x):
    return 0.6*x**2 + 1.5
def t(x):
    return 0.8 * x**3 -10*x +10

x = np.arange(-10, 10, 0.1)
plt.plot(x, w(x))#, label = 'w(x)') #azul
plt.plot(x, t(x))#, label = 't(x)')"naranja"
plt.title('Nombre del gráfico')
plt.ylabel('Nombre eje y')
plt.xlabel('Nombre eje x')
plt.xlim(-3,4)
plt.ylim(-4,24)
plt.legend()
plt.grid()
#opcional:
ax=plt.gca()
ax.spines[["left","bottom"]].set_position(("data",0))
ax.spines[["right","top"]].set_visible(False)
#

plt.show()
```

```python
#determine el intervalo donde w(x) es mayor o igual a t(x)
#  't(x)')"naranja"/'w(x)') #azul
import sympy as sp
def w(x):
    return 0.6*x**2 + 1.5
def t(x):
    return 0.8 * x**3 -10*x +10

x=sp.Symbol("x")
ecuacion = w(x)-t(x) 
solucion=sp.solve(ecuacion,x)
x1 = round(solucion[0],1)
x2 = round(solucion[1],1)
x3 = round(solucion[2],1)
print(f'las funciones se intersecan en los puntos {x1,round(w(x1),1)} y {x2, round(w(x2))} y {x3, round(w(x3),1)}')
print(f'La funcion w(x) es mayor en el punto -6 hasta el punto {x1} luego es mayor desde el punto {x2} hasta el punto  ')

```
