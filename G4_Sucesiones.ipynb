Definir que tipo de susecion es esta y porque:

R = 0.9 Valor de reduccion o aumento
"Inicial" = Q0 = 3


a_1 = a_0*R
Formula o término general: a_n = a_1*R**(n-1)

```python
a_0 = 3
R = 0.9#100% - 10% = 90% = 0.9
a_1 = a_0 * R

def a(n):
    return a_1*R**(n-1)

print((a(10)))
print(f'La quita fase duró {round(a(5))} meses aproximadamente')

lista = []
for i in range (0,11):
    lista.append(a(i))
suma = round(sum(lista),1)
print(f'la duracion del proyecto luego de las 10 semanas es de: {suma}')
```

# Problema 12

Una empresa de publicidad ofrece para un puesto de trabajo un salario inicial y un aumento mensual en una razón constante. Javier se presenta a la entrevista, y el encargado del departamento de recursos humanos le explica que en el segundo mes de trabajo su sueldo será de $\$402.000$, y que en el tercer mes de trabajo será de $\$404.010$. Determine:

1. El salario aproximado que Javier recibiría en el décimo mes de trabajo en la empresa.
2. ¿Cuál sería el ingreso total recibido por los sueldos de Javier durante el tercer año de trabajo?

```python
#Sucesion geometrica ya que presenta una razon constante

a_2 = 402000
a_3 = 404010
R = (a_3 / a_2) ** (1/1)
a_1 = a_3*R**(-2)

def A(n):
    return a_1*R**(n-1)
print(f'Al decimo mes, el salario aproximado que javier recibirá, será de: ${round(A(10))} pesos')

total = []
for i in range (25,37):
    total.append(A(i))
suma = round(sum(total))
print(f'Durante el tercer año recibe en total: ${suma} Pesos')


```

# Problema 14

Al llenar un tambor con agua, se registra que durante el minuto $4$ el tambor recibe $128$ litros, y durante el minuto $9$ recibe $88$ litros. Sabiendo que en cada minuto recibe una cantidad $d$ de litros de agua menos que en el minuto anterior, determine:

1. ¿En qué instante el tambor deja de recibir agua?
2. ¿Cuántos litros recibe en total el tambor?

Realice los cálculos anteriores utilizando *Python*.

```python
a_4 = 128
a_9 = 88
d = (a_9-a_4)/5 #resto el termino de posicion mayor dividido en la diferencia
a_1 = a_9 - 8 * d

def C(n):
    return a_1+(n-1)*d

import sympy as sp
n = sp.Symbol('n')
ecuacion = C(n) - 0
solucion = sp.solve(ecuacion,n)
print(f'En el minuto {solucion} el tambor deja de recibir agua')
total=[]
for i in range(1,21):
    total.append(C(i))
suma = sum(total)
print(f'El tambor recibe {suma} Litros de agua')
```
