# Problema 3

Las filas de la matriz $M$, en el orden mencionado, nos muestran la cantidad de transistores y resistores que se requieren por cada artículo, según cada modelo. Por otra parte, las filas de la matriz $Q$ nos reportan las cantidades producidas de cada modelo para las tres primeras semanas del mes pasado.

<br>

\begin{split}
M = \begin{pmatrix}
 9 &  14 \\
 4 &  5  \\
\end{pmatrix} \ \ \ \ \ \ \ \ Q = \begin{pmatrix}
 200 &  240 & 220 \\
 175 &  210 & 215 \\
\end{pmatrix}
\end{split}

<br>

Además, considere las matrices $G = M \cdot Q$ y $H = Q \cdot M$.

1. Indique el orden de $M$ y $Q$.
2. ¿Es posible calcular las matrices $G$ y $H$? Justifique.
3. Determine utilizando *Python* la matriz $G$ e interprete los elementos $g_{23}$  y $g_{12}$.
4. ¿Cuántos resistores se utilizaron durante las tres semanas?

```python
import numpy as np
M = np.array([[9,14],[4,5]])
Q = np.array([[200,240,220],[175,210,215]])
#Pregunta1: orden de M = 2x2 Q = 2x3
#pregunta2:G2x3 = M2x2*Q2x3 es posible porque 
#la cantidad de columna de M es igual a la cantidad de filas de Q

#   H = Q2x3 * M2x2 No es posible ya que la cantidad de columna de Q no es igual de la cantidad de fila que M
#pregunta3 calculandoG, G=M*Q (NO SE HACE)
G = M.dot(Q) #ASI SE MULTIPLICAN MATRICES CON PYTHON
print(f'La matrixg=\n{G}')
g23 = G[2-1,3-1] #fila2 los resistores:  Columna 3 semana 3
print(f'se utilizo en total, {g23} resistores en la semana 3 (En ambos modelos)')
g12 = G[1-1,2-1] #fila1: transisotres; columna 2; semana2
print(f'Se utilizo en total {g12} transistores en la semana 2(En la produccion de ambos modelos)')
#pregunta 4: 
gs3 = np.sum(G,axis=1)# asu tambien, G[1].sum() para sumar elementos de una fila
print(f'En total se utlizaron {gs3[1]} resistores las 3 semanas')
```

2. Considera que el siguiente sistema de ecuaciones:

<br>

\begin{split}
h + m & = 60 \\
0,16h + 0,2m & = 11
\end{split}

<br>

$\hspace{1cm}$ Se puede escribir matricialmente como:

<br>

\begin{split}
\begin{pmatrix}
 1  &  1 \\
 0,16 & 0,2 \\
\end{pmatrix} \cdot \begin{pmatrix}
 h \\
 m \\
\end{pmatrix} = \begin{pmatrix}
 60 \\
 11 \\
\end{pmatrix}
\end{split}

<br>

* Utiliza la matriz inversa para encontrar los valores de $h$ y $m$ que permiten satisfacer el sistema de ecuaciones.

```python
A = np.array([[2,1],[3,-2]])
B=np.array([[7],[21]])
A_inv = np.linalg.inv(A)
X = A_inv.dot(B)
print(X)
#Comprobando
B_comp = A.dot(X)
print(B_comp)
```

```python
A = np.array([[1,1],[0.16,0.2]])
B=np.array([[60],[11]])
A_inv=np.linalg.inv(A)
H=A_inv.dot(B)
print(H)
#comprobando
B_comp=A.dot(H)
print(B_comp)
```
