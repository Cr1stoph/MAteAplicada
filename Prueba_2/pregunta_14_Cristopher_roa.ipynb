```python
import numpy as np
A = np.array([[700,1400],[1,1]])
B = np.array([[70000],[80]])
A_inv = np.linalg.inv(A)
R = A_inv.dot(B)
print(R)
print(700*60 + 1400*20)
print(60+20)
print(f'La cantidad de conectores R45cat5 que se deben vender para asegurar la rentabilidad mínima es de {R[0][0]}')
print(f'Para los conectores R45cat6 se deben vender {R[1][0]} para asegurar la rentabilidad mínima')
```

- La cantidad de conectores R45cat5 que se deben vender para asegurar la rentabilidad mínima es de 60 
- Para los conectores R45cat6 se deben vender 20 para asegurar la rentabilidad mínima.
