```python
#El tamaño de una base de datos en terabytes (TB) crece exponencialmente en función del tiempo en años, a través del siguiente modelo:
#﻿D left parenthesis t right parenthesis equals 2 times 1 comma 2 to the power of t﻿
#¿Cuántos años deben pasar que la base de datos tenga un tamaño de 20 TB?
def D(t):
    return 2*1.2**(t)
import sympy as sp
x = sp.Symbol('x')
ecuacion = D(x)-20
solucion = sp.solve(ecuacion,x)
print(solucion)
print(D(solucion[0]))
#Esto para resolver una ecuacion.
```

```python
# A 15x10 filas: Inversionista
#       columnas: tipo Accion
# B 10x5 filas: tipo Accion
#       columnas: Dia habil
# R15x5 si se multiplican R = resultante
# R15x5 Inversionistas x dias
import numpy as np
#Cración de matrices mediante función sobre los índices

A = np.fromfunction(lambda i, j: 20*(i+1)*(j+1), (15,10))
E = np.fromfunction(lambda i, j: 4*(i+1)+(j+1), (10,5))
print(A)
print(E)
#En cuanto está evaluando, para el miercoles de la semana pasasa,
#El total de acciones que posee en esta empresa el inversionista 8
R= A.dot(E)
print(R[8-1,3-1])
```
