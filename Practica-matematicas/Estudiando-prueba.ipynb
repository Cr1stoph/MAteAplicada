$$Problema 2$$
- Sea B_n=2n^3-12n^2, determine el valor de n para que B_n sea 13.284.800
- Sea C_n=0,5n^3-0.2n, determine el valor de n para que C_n sea 58.814,7


```python
def B(n):
    return 2*(n**3)-12*(n**2)

import sympy as sp
x = sp.Symbol('x')
ecuacion = B(x) - 13284800
solucion = sp.solve(ecuacion,x)
print(solucion[0])
print(solucion[1])
print(solucion[2])
print(f'13284800 se encuentra en la posición: {solucion[0]}')
print(f'Es decir que n es: {solucion[0]}')
print('-----------------------------------------')

def C(n):
    return 0.5*(n**3) - 0.2*n

y = sp.Symbol('y')
ecuacion1 = C(y) - 58814.7
solucion1 = sp.solve(ecuacion1,y)
print(solucion1[0])
print(solucion1[1])
print(solucion1[2])
print(f'58814.7 se encuentra en la posicion {solucion1[0]}')
print(f'"n" es es: {solucion1[0]}')
```

$$Problema 3$$
Un atleta está entrenando para un maratón y aumenta la distancia que corre cada semana en 1.5 km respecto a la distancia de la semana anterior. Si en la primera semana corrió 5 km, 

- ¿Qué tipo de sucesión modela la situación, justifique
- ¿cuál es la expresión permite calcular la distancia que corre en la semana n?


```python
#La sucesion que modela la situación es una sucesion aritmética.
#La expresion que permite calcular la distancia que corre en la semana n:
# a_n = a_1+(n-1)*d 
d = 1.5 #km
a_1 = 5 #km
def A(n):
    return a_1 + (n-1)*d
```

$$Problema 4$$
Un automóvil de colección inicialmente está valorizado en 1500 euros. Según los tasadores el precio del automóvil se irá duplicando cada año. 
- ¿Qué tipo de sucesión modela al precio del automóvil?
- Exprese el valor del automóvil al cabo de n años, en términos del valor al cabo de 1 año.


```python
#La sicesión que modela al precio del automovil es una suscesión geometrica
a_1 = 1500
a_2 = 3000
r = (a_2/a_1)**1/1
def A(n):
    return a_1 * r**(n-1) 
lista = []
for i in range (1,13):
    lista.append(A(i))
print(f'Al final del año, el vehiculo estará costando: {lista[-1]}')
```

$$Problema 5$$
Un depósito de agua tiene una fuga constante que reduce su nivel en 8 litros por cada hora. Si al inicialmente el depósito contenía 5000 litros.
- ¿Qué tipo de sucesión modela la situación?
- ¿Cuál es la expresión que permite calcular la cantidad de agua en el depósito transcurridas n horas, en términos de la cantidad inicial de litros?, y ¿en términos de la cantidad al cabo de una hora?
- ¿Cuántos litros de agua quedarán al cabo 1 día?


```python
#El tipo de sicesión que modela la situación es una sucesion aritmetica
a_0 =5000#Litros
d = -8
def A(n):
    return a_0+(n)*d
lista = []
for i in range (0,25):
    lista.append(A(i))
print(f'Transcurrido una hora el deposito tendrá: {lista[1]}L')
print(f'Al cabo de un día, el depósito, tendrá: {lista[-1]}L')
```

$$Problema 6$$
Un grupo de seguidores en una red social comienza a duplicarse cada 6 meses debido a una campaña publicitaria. Si inicialmente cuenta con 150 seguidores. 
- ¿Cuántos seguidores tendrá al cabo de 2 años?


```python
#Sucesión geometrica
a_0 = 150 #seguidores el mes 0
a_6 = 300 #seguires al mes 6
r = (a_6/a_0)**(1/6)
def A(n):
    return a_0 *(r**n)
print(f'Al cabo de dos años, la cuenta de instagram temdrá: {round(A(24))} Seguidores')
```

$$Problema 7$$

- ¿Cuántos dispositivos estarán infectados al cabo de 7 semanas?


```python
a_1 = 3
r= 2
def A(n):
    return a_1*r**(n-1)
print(f'Al cabo de 7 semanas, estarán {A(7)}')
```

$$Problema 10$$
Las matrices P y Q representan la cantidad de productos vendidos en dos tiendas durante dos días.
- La matriz P corresponde a la tienda 1:  
Las filas representan cuatro tipos de frutas (manzanas, plátanos, naranjas, uvas) y las columnas representan el día lunes y martes, en ese orden.
- La matriz Q corresponde a la tienda 2:  
con la misma correspondencia de filas y columnas.
- Si R es la matriz resultante de P+Q, interprete el elemento r_31.
- ¿Es posible realizar P⋅ Q?



```python
P = np.array([[15,20],[8,12],[22,10],[5,18]])
Q =  np.array([[10,25],[7,13],[18,9],[6,10]])
R = P+Q
print(R)
r_31 = R[3-1,1-1]
print(f'En total en las dos tiendas se han vendido {r_31} naranjas')
```
