***$$\color{red}{\text{Término general de una sucesión aritmética:}}$$***

**#Dependiendo del $a_1$**
$$\mathbf{ a_{n} = a_{1} + (n - 1)\cdot d}$$

**#Dependiendo del $a_0$:** $\quad a_{n} = a_{0} + (n)\cdot d$

Si $\textbf{n > k:}$     $\quad d=\dfrac{a_{n}-a_{k}}{n-k}$

***$$\color{red}{\text{Término general de una sucesión geométrica:}}$$***


**#Dependiendo del $a_1$**
$$\mathbf{a_{n} = a_{1} \cdot r^{n-1}}$$
**#Dependiendo del $a_0$:** $\quad a_{n} = a_{0} \cdot r^{n}$

Si $\textbf{n > k:}$      $\quad r=\sqrt[n-k]{\dfrac{a_{n}}{a_k}}$
