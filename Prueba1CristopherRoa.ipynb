$$Pregunta 5$$

```python
def A(t):
    return 0.2 * (t**2) + 3 * t + 5.2

def B(t):
    return 0.2*(t**2) + 4 * t + 3.2
import sympy as sp
x = sp.Symbol('x')
ecuacion = A(x) - B(x)
solucion = sp.solve(ecuacion,x)
print()
```


$$Pregunta 8$$

```python
import numpy as np
e = np.e

def S(t):
    return 1000 * (1-e**-0.3*t)
print(S(0.764020))
print(S(0.8))
import sympy as sp
x = sp.Symbol('x')
ecuacion = S(x) - 434
solucion = sp.solve(ecuacion,x)
print(round(solucion[0],1))
```

```python
def f(x): #Sitio 1
    return 335 * x
def g(x):
    return 245 * x + 180

import matplotlib.pyplot as plt
x = np.arange(1, 11, 1)
plt.plot(x, f(x), label = 'f(x) sitio 1')
plt.plot(x, g(x), label = 'g(x) sitio 2')
plt.title('Precio de tablets en funcion de cada sitio de internet')
plt.ylabel('Precio (dolares)')
plt.xlabel('Cantidad tablets (unidades)')
plt.legend()
plt.grid()

#opcional:
ax=plt.gca()
ax.spines[["left","bottom"]].set_position(("data",0))
ax.spines[["right","top"]].set_visible(False)
```
