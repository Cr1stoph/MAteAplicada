# Problema 1

Un equipo de desarrollo está evaluando la eficiencia de un nuevo algoritmo implementado en *Python*. Han medido el tiempo de ejecución del algoritmo en función de la cantidad de elementos de entrada, obteniendo los resultados que se muestran en la siguiente tabla:

<p><center>

| Cantidad de elementos (unidades) | Tiempo de ejecución (ms) |
|----------------------------------|--------------------------|
|$$ 100 $$                         | $$2 $$                   |
|$$ 200 $$                         | $$4 $$                   |
|$$ 500 $$                         | $$10 $$                  |
|$$ 1.000 $$                       | $$20 $$                  |
|$$ 2.000 $$                       | $$40 $$                  |

</center></p>

1. Defina variable dependiente e independiente, indicando unidad de medida.
- Variable dependiente: tiempo de ejecucion (ms)
- Variable independiente: Cantidad de elementos(unidades)
2. Encuentra la función que relaciona el tamaño de la entrada con el tiempo de ejecución del algoritmo, para esto utiliza la función *polyfit* de *Numpy*.
3. Calcula el tiempo de ejecución para una entrada de $1.500$ elementos.
4. Utiliza la función *fsolve* de *Scipy* para determinar el tamaño de entrada que haría que el tiempo de ejecución sea de $50$ milisegundos.


```python
import numpy as np
#la funcon plyfit depende de 3 argumentos
#la vamos a llamar con el np.polyfit(1,2,3) el 1 será una lista que tendrá elementos de la variable independiente
#2 será una lista que tendrá los elementos que contenga la variable dependiente
#3 tendrá el grado del polinomio
#--------------obtener m y n-----------------------
print('---------------Obtener m y n---------------')
vi = [100,200,500,1000,2000]
vd = [2,4,10,20,40]
coeficiente = np.polyfit(vi,vd,1)
print(np.around(coeficiente,5)) #primero pendiente y segundo coeficiente de posicion

#-----------FORMA MEJOR DE HACER-------------------- 
print('---------------MEJOR FORMA DE HACER---------------')

m1,n1 = np.polyfit(vi,vd,1)
m = round(m1,5)
n = round(n1,5)
print(m,n )#recordar que la forma algebraica de la funcoin lineal es f(x) = mx +n
print(f'la forma algebraica es: f(x) = {m}x + {n}') #ya que n es 0 no es necesario escribirlo

#-------------funcion-------------------------------

def f(elementos): #e de elementos
    return 0.02 *elementos
print('Comprobacion')
print(f(1000)) #se comprueba

#-------------Evaluar funcion para pregunta 3----------------
print('---------------Pregunta 3---------------\n')
#pregunta 3
print(f'Con las 1500 elementos serán {f(1500)} ms\n')

#--------------------Pregunta 4-----------------------
print('---------------Pregunta 4---------------\n')
import sympy as sp

x = sp.Symbol('x')
ecuacion = f(x) - 50
solucion = sp.solve(ecuacion,x)
print(solucion[0])
print('\n--------------interpretacion--------------\n')
print(f'para que la ejecucion sea de 50ms tendrán que ser {round(solucion[0])} elementos')
print(f'La pendiente {m} por cada 1 elemento aumenta en 0,02ms')
print(f'm = {m} El incremento de tiempo 0,02ms  por el incremento de 1 elemento')
print('funcion lineal tiene dos parametros pendiente y coefiiente de posicion')
print('polifit depende de la variable indep, variable dependiente')
```
