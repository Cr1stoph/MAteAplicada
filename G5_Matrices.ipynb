# Problema 1

Las matrices $A$ y $B$ representan las ventas en dos años, en millones de dólares, de los tres productos de una empresa en las cuatro regiones en que está disponible. Es decir, $a_{ij}$ y $b_{ij}$ representan el monto de las ventas del producto $i$ en la región $j$ el primer y segundo año, respectivamente:

<br>

\begin{split}
A = \begin{pmatrix}
 2,6 &  4,8 &  1,8 & 0,9 \\
 3,2 &  4,4 &  2,5 & 2,8 \\
 2,4 &  3,6 &  3,8 & 2,5 \\
\end{pmatrix} \text{ } B= \begin{pmatrix}
 3,6 &  2,5 &  3,0 & 2,5 \\
 4,5 &  5,0 &  3,5 & 3,8 \\
 2,9 &  3,0 &  4,6 & 4,0 \\
\end{pmatrix}
\end{split}

<br>

1. Indique el orden de cada matriz.
2. Calcule la matriz $S = A + B$.
3. Calcule la matriz $D = B - A$.
4. Indique el valor e interprete los elementos $s_{24}$ y $d_{24}$.
5. Para el tercer año de funcionamiento, los tres productos ahora están disponibles en seis regiones. Durante este año, las ventas de esta empresa estarán registradas en una matriz $E$. ¿Es posible determinar la matriz $T = A + B + E$? Justifique su respuesta.

```python
#Pregunta 1: el orden de ambas matrices es de 3x4 3 filas y 4 columnas

import numpy as np
A = np.array([[2.6, 4.8, 1.8, 0.9],[3.2, 4.4, 2.5, 2.8],[2.4, 3.6, 3.8, 2.5]])

B = np.array([[3.6, 2.5, 3.0, 2.5],[4.5, 5.0, 3.5, 3.8],[2.9, 3.0, 4.6, 4.0]])
#Pregunta2
S = A + B
print(f'S= {S}.')
#Pregunta 3
D = B-A
print(f'D = {D}.')
#4. Indique el valor e interprete los elementos de "s_24" y "d_24"
s_24 =S[2-1][4-1] #Fila 2: producto2 ; Columna 4: region 4. es la 4-1 ya que comienza en 0 python
d_24 = D [2-1][4-1]
print(d_24) #d_24=1 
print(f'El total de ventas, del producto 2 en la región 4 fue de {s_24}'\
    'millones de dólares, entre ambos años')
print(f'La diferenia entre la veta entre segundo y primer año del producto 2 en la region 4'\
    'fue de {d_24} millones de dolares')
#5.
print('No, no es posible determinar la matriz T = A + B + E ya que las matrices no tienen el mismo ordem')
```
